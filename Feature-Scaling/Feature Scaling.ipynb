```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv(r"C:\Users\LENOVO\Downloads\European_Ski_Resorts.csv",encoding='latin1')
dataset.head(2)
```

```python
dataset.isnull().sum()
```

```python
sns.distplot(dataset["HighestPoint"])
plt.show()
```

```python
dataset.describe()
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss.fit(dataset[["HighestPoint"]])
```

```python
dataset["HighestPoint_ss"]=pd.DataFrame(ss.transform(dataset[["HighestPoint"]]),columns=["x"])
```

```python
dataset.head(3)
```

```python
dataset.describe()
```

```python

plt.figure(figsize=(10,7))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["HighestPoint"])
plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["HighestPoint_ss"])
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv(r"C:\Users\LENOVO\Downloads\diabetes.csv",encoding='latin1')
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.distplot(dataset["SkinThickness"])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
ms= MinMaxScaler()
ms.fit(dataset[["SkinThickness"]])
```

```python
dataset["SkinThickness_min"]=ms.transform(dataset[["SkinThickness"]])
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Before")
sns.distplot(dataset["SkinThickness"])

plt.subplot(1,2,2)
plt.title("After")
sns.distplot(dataset["SkinThickness_min"])
plt.show()
```
