```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
```

```python
dataset=pd.read_csv(r"C:\Users\LENOVO\Downloads\placement.csv")
```

```python
dataset.head(3)
```

```python
plt.figure(figsize=(5,3))
sns.scatterplot(x="cgpa",y="package",data=dataset)
plt.show()
```

```python
dataset.isnull().sum()
```

```python
x=dataset[["cgpa"]]
y=dataset["package"]
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#training process
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
#y=mx+c
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
#y=0.57425647*x-1.02700693745421
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[6.89]])
```

```python
0.57425647*6.89-1.02700693745421
```

```python
# checking the predicted line
```

```python
y_prd=lr.predict(x)
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(x="cgpa",y="package",data=dataset)
plt.plot(dataset["cgpa"],y_prd,c="red")

plt.legend(["org data","predict line"])
plt.savefig("predict.jpg")
plt.show()
```

```python
## POLYNOMIAL REGRESSION
```

```python
import pandas as pd 
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv(r"C:\Users\LENOVO\Downloads\salary_level.csv")
```

```python
dataset.head(3)
```

```python
dataset.corr()
```

```python
plt.scatter(dataset["Level "],dataset["Salary"])
plt.xlabel("Level")
plt.ylabel("Salary")
plt.show()
```

```python
x=dataset[["Level "]]
y=dataset["Salary"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf=PolynomialFeatures(degree=2)
pf.fit(x)
x=pf.transform(x)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LinearRegression
```

```python
#training process
lr=LinearRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
#y=m1x1+m2x2+c
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
prd=lr.predict(x)
```

```python
plt.scatter(dataset["Level "],dataset["Salary"])
plt.plot(dataset["Level "],prd,c='red')
plt.xlabel("Level")
plt.ylabel("Salary")
plt.legend(["org","red"])
plt.show()
```

```python
test=pf.transform([[8]])
test
```
