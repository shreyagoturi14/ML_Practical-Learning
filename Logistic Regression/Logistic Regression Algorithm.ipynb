## Logistic regression single input

```python
import pandas as pd 
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
dataset=pd.read_csv(r"C:\Users\LENOVO\Downloads\Social_Network_Ads.csv")
dataset.drop(columns="EstimatedSalary",inplace=True)
```

dataset.head(3)

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x="Age",y="Purchased",data=dataset,color="red")
plt.show()
```

```python
x=dataset[["Age"]]
y=dataset["Purchased"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[10]])
```

```python
#finding the predict line 
plt.figure(figsize=(4,3))
sns.scatterplot(x="Age",y="Purchased",data=dataset)
sns.lineplot(x="Age",y=lr.predict(x),data=dataset,color='red')
plt.show()
```

 Logistic Regression - Multiple Input


```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv(r"C:\Users\LENOVO\Downloads\placement (2).csv")
```

```python
dataset.head(3)
```

```python
dataset.drop(columns="Unnamed: 0",inplace=True)
```

```python
dataset.head(3)
```

```python
sns.scatterplot(x="cgpa",y="iq",data=dataset,hue="placement")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["placement"]
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```

```python
lr.predict([[5.9,106.0]])
```

```python
lr.coef_
```

```python
lr.intercept_
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=lr)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset=pd.read_csv(r"C:\Users\LENOVO\Downloads\placement (1).csv")
```

```python
dataset.head(3)
```

```python
sns.scatterplot(x="cgpa",y="placement_exam_marks",data=dataset,hue="placed")
plt.show()
```

```python
x=dataset.iloc[:,:-1]
y=dataset["placed"]
```

```python
from sklearn.preprocessing import PolynomialFeatures
```

```python
pf=PolynomialFeatures(degree=2)
pf.fit(x)
x=pd.DataFrame(pf.transform(x))
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train,y_train)
```

```python
lr.score(x_test,y_test)*100
```
