```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
dataset = pd.read_csv(r"C:\\Users\\LENOVO\\Downloads\\European_Ski_Resorts.csv",encoding='latin1')
dataset.head(4)
```

```python
dataset.info()
```

```python
dataset.describe()
```

```python
plt.figure(figsize=(10,5))
sns.boxplot(x="BeginnerSlope",data=dataset)
plt.show()
```

```python
sns.displot(dataset["BeginnerSlope"],kde=True)
plt.show()
```

```python
#IQR methods in removing the outlier
```

```python
dataset.shape
```

```python
q1=dataset["BeginnerSlope"].quantile(0.25)
```

```python
q3=dataset["BeginnerSlope"].quantile(0.75)
```

```python
IQR=q3-q1
```

```python
min_range=q1-(1.5*IQR)
max_range=q3+(1.5*IQR)
min_range,max_range
```

```python
new_dataset=dataset[dataset["BeginnerSlope"]<=max_range]
```

```python
new_dataset.shape
```

```python
sns.boxplot(x="BeginnerSlope",data=new_dataset)
plt.show()
```

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt 
```

```python
dataset = pd.read_csv(r"C:\\Users\\LENOVO\\Downloads\\diabetes.csv",encoding='latin1')
dataset.head(4)
```

```python
dataset.isnull().sum()
```

```python
dataset.describe()
```

```python
sns.boxplot(x="Age",data=dataset)
```

```python
sns.displot(dataset["Age"],kde=True)
```

```python
min_range=dataset["Age"].mean()-(3*dataset["Age"].std())
```

```python
max_range=dataset["Age"].mean()+(3*dataset["Age"].std())
```

```python
min_range,max_range
```

```python
new_data=dataset[dataset["Age"]<=max_range]
```

```python
sns.boxplot(x="Age",data=new_data)
```

```python
Explored statistical outlier detection using Z-score.
```

```python
z_score=(dataset["Age"]-dataset["Age"].mean())/(dataset["Age"].std())
```

```python
z_score>3
```

```python
dataset["z_score"]=z_score
```

```python
dataset
```
